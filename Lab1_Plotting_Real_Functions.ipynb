## Aufgabe 1

1. Es sei f : R → R, f(x) = x^2/(x^2 + 1) .Man stelle f dar, wobei folgende Fragen ber¨ucksichtigt werden.

a) Was sind passende untere und obere Grenzen f¨ur die die dargestellten Intervalle auf den
Ox bzw. Oy Achsen? (Vergleiche entstehende Bilder mit x ∈ [−1, 1], y ∈ [−1, 1] und mit x ∈
[−100, 100], y ∈ [−1, 1].)

b) L¨asst f horizontale Asymptoten zu? (Falls ja, sollten auch diese im selben Bild wie f
dargestellt werde.)

c) Das Taylor Polynom zweiten Grades welches f um x0 = 0 approximiert ist T(x) =? (Stelle
auch diese Polynomialfunktion im selben Bild dar.)

### Lösungsidee

Die Funktion ist


f(x)=x^2/(x^2+1).


Da x^2 >= 0 und  x^2+1 > 0  fur alle x ∈ R, gilt


0 <=  f(x) < 1.


Deshalb ist fur die y-Achse ein Bereich wie zum Beispiel

[-0.1, 1.1]

gut geeignet.

Fur die x-Achse vergleichen wir einen kleinen Bereich [-1,1] mit einem großen Bereich [-100,100], um das lokale und globale Verhalten der Funktion zu sehen.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2 / (x**2 + 1)

x1 = np.linspace(-1, 1, 1000)
x2 = np.linspace(-100, 100, 4000)

y1 = f(x1)
y2 = f(x2)

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].plot(x1, y1, label=r"$f(x)= x^2/(x^2+1)$", linewidth=2)
axes[0].set_title("Comparison on x ∈ [-1, 1]")
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")
axes[0].set_xlim(-1, 1)
axes[0].set_ylim(-0.1, 1.1)
axes[0].grid(True)
axes[0].legend()

axes[1].plot(x2, y2, label=r"$f(x)= x^2/(x^2+1)$", linewidth=2)
axes[1].set_title("Comparison on x ∈ [-100, 100]")
axes[1].set_xlabel("x")
axes[1].set_ylabel("y")
axes[1].set_xlim(-100, 100)
axes[1].set_ylim(-0.1, 1.1)
axes[1].grid(True)
axes[1].legend()

plt.tight_layout()
plt.show()
```

### a)
Fur kleine x-Werte zeigt das Intervall [-1,1] den Verlauf in der Nahe von 0.
Fur große x-Werte zeigt das Intervall [-100,100], dass sich die Funktion einem festen Wert annahert.

Da stets 0 <= f(x) < 1


gilt, ist ein sinnvoller Bereich fur die y-Achse ungefahr [-0.1, 1.1].

### b) Horizontale Asymptote

Wir untersuchen den Grenzwert für x -> +/- infty:

lim(x -> +/- infty)[x^2/(x^2+1)] = 1

Also besitzt die Funktion die horizontale Asymptote y=1.

```python
x = np.linspace(-25, 25, 2000)
y = f(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label=r"$f(x)=x^2/(x^2+1)$", linewidth=2)
plt.axhline(y=1, color="red", linestyle="--", label="horizontal asymptote y = 1")
plt.title("Function with horizontal asymptote")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-25, 25)
plt.ylim(-0.1, 1.1)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

### c) Taylor-Polynom zweiten Grades um x0=0

Wir schreiben

f(x)=x^2/(1+x^2).

Mit der Reihenentwicklung

1/(1+x^2)=1-x^2+x^4- ...

folgt

f(x)=x^2(1-x^2+x^4-...)=x^2-x^4+x^6-...

Damit ist das Taylor-Polynom zweiten Grades um \( x0=0 \):

T2(x)=x^2.

```python
def taylor_2(x):
    return x**2

x = np.linspace(-2, 2, 2000)
y_f = f(x)
y_t = taylor_2(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y_f, label=r"$f(x)=\frac{x^2}{x^2+1}$", linewidth=2)
plt.plot(x, y_t, label=r"$T_2(x)=x^2$", linewidth=2)
plt.axhline(y=1, color="red", linestyle="--", label="horizontal asymptote y = 1")
plt.title("Function, asymptote, and Taylor polynomial of degree 2")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-2, 2)
plt.ylim(-0.1, 4.1)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## Schlussfolgerung

- Die Funktion f(x)=x^2/(x^2+1) liegt fur alle x ∈ R im Bereich

0 <= f(x) < 1

- Eine horizontale Asymptote ist vorhanden, namlich y=1.

- Das Taylor-Polynom zweiten Grades um x0=0 ist

T2(x)=x^2.

- In der Darstellung sieht man, dass T2(x) die Funktion nur in der Nahe von x=0 gut approximiert.

## Eigenschaften der Rosenbrock Funktion

Die Rosenbrock Funktion ist eine bekannte Testfunktion für Optimierungsalgorithmen.

Sie besitzt:

- ein globales Minimum
- eine lange, schmale und gekrummte Talstruktur

Das Minimum liegt bei

(x,y) = (1,1) und f(1,1) = 0

Diese Struktur macht es fur viele Optimierungsalgorithmen schwierig, das Minimum zu finden.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

def rosenbrock(x, y):
    return (1 - x)**2 + 100 * (y - x**2)**2

x = np.linspace(-2, 2, 400)
y = np.linspace(-1, 3, 400)

X, Y = np.meshgrid(x, y)
Z = rosenbrock(X, Y)
```

## Darstellung der Funktion

Wir stellen die Funktion dar als

1. 3D Oberflache
2. Contour Plot

Die betrachtete Domane ist x∈[−2,2], y∈[−1,3]

```python
fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap=cm.viridis, linewidth=0)

ax.set_title("Rosenbrock Function - 3D Surface")
ax.set_xlabel("x")
ax.set_ylabel("y")

fig.colorbar(surf)

plt.show()
```

```python
plt.figure(figsize=(7,5))

levels = np.geomspace(1e-2, Z.max(), 30)

contours = plt.contour(X, Y, Z, levels=levels)

plt.clabel(contours, inline=True, fontsize=8)

plt.plot(1,1,"r*",markersize=12,label="global minimum (1,1)")

plt.title("Rosenbrock Function - Contour Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.show()
```

## Minimum der Funktion

Die Rosenbrock-Funktion lautet

f(x, y) = (1-x)^2 + 100(y-x^2)^2

Da beide Terme Quadrate sind, gilt fru alle (x, y) ∈ R^2

f(x, y) >= 0

Die Funktion ist also nach unter durch 0 beschrankt.
Der kleinste mogliche Funktionswert ist daher 0. Dieser Wert wird genau dann angenommen, wenn beide Quadratterme gleich 0 sind, also wenn:

(1 - x)^2 = 0  und (y - x^2)^2 = 0

Aus der ersten Gleichung folgt:

x = 1

Setzt man dies in die zweite Gleichung ein, so erhalt man:

y = x^2 = 1

Damit wird der Wert 0 nur im Punkt:

(x, y) = (1, 1) angenommen.

Also besitzt die Rosenbrock-Funktion ein eindeutiges globales Minimum bei

(1, 1) mit f(1, 1) = 0.

Dies lasst sich zusatzlich direkt uberprufen:

```python
rosenbrock(1,1)
```

## Untersuchung der Konvexitat

Eine zweimal differenzierbare Funktion ist konvex, wenn ihre Hesse Matrix uberall positiv semidefinit ist.

Die Hesse Matrix der Rosenbrock Funktion ist H(x,y) =
\begin{pmatrix}
2 - 400y + 1200x^2 & -400x \\
-400x & 200
\end{pmatrix}

Wenn diese Matrix irgendwo negative Eigenwerte besitzt, ist die Funktion nicht konvex.

```python
def hessian_rosenbrock(x, y):

    f_xx = 2 - 400*y + 1200*x**2
    f_xy = -400*x
    f_yy = 200

    return np.array([[f_xx, f_xy],
                     [f_xy, f_yy]])
```

```python
H = hessian_rosenbrock(1, 1)

print("Hessian matrix H(1,1) =")
print(H)
```

```python
grid = 100

xs = np.linspace(-2,2,grid)
ys = np.linspace(-1,3,grid)

min_eig = []

for x in xs:
    for y in ys:
        H = hessian_rosenbrock(x,y)
        eigvals = np.linalg.eigvalsh(H)
        min_eig.append(eigvals.min())

print("Smallest sampled eigenvalue:", min(min_eig))
```

## Ergebnis zur Konvexitat

Da die kleinste Eigenwert der Hesse Matrix in unserem Bereich negativ wird, ist die Hesse Matrix nicht überall positiv semidefinit.

 - Damit ist die Rosenbrock Funktion nicht global konvex.

 - Damit sind nicht alle Subniveaumengen konvex.

## Warum wird die Rosenbrock Funktion als Testfunktion benutzt?

Die Rosenbrock Funktion ist schwierig fur viele Optimierungsalgorithmen, weil:

- sie ein sehr schmales Tal besitzt
- dieses Tal gekrummt ist
- der Gradient oft nicht direkt zum Minimum zeigt

Dadurch mussen Optimierungsalgorithmen sehr sorgfaltig entlang des Tals navigieren.

Aus diesem Grund wird die Rosenbrock Funktion haufig verwendet, um die Leistungsfahigkeit von Optimierungsalgorithmen zu testen.
